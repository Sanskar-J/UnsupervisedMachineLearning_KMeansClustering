## Trying out K Means Clustring with 3 clusters

```python
import sys
print(sys.version)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate Data- two means first
mean1=[np.random.randint(100),np.random.randint(100)]
mean2=[np.random.randint(100),np.random.randint(100)]
mean3=[np.random.randint(150),np.random.randint(150)]

# make diagonal covariance
cov=[[100,0],[0,100]]
x1,y1=np.random.multivariate_normal(mean1,cov,100).T
x2,y2=np.random.multivariate_normal(mean2,cov,100).T
x3,y3=np.random.multivariate_normal(mean3,cov,100).T

x=np.append(x1,x2)
x=np.append(x,x3)
y=np.append(y1,y2)
y=np.append(y,y3)

plt.plot(x,y,'x')
plt.axis('equal')
plt.show()
```

```python
X=np.array(list(zip(x,y)))
```

```python
# making kmeans model
kmeans=KMeans(n_clusters=3)
```

```python
# fit the model to the data
kmeans.fit(X)
```

```python
centroids=kmeans.cluster_centers_
labels=kmeans.labels_

print(labels)
print(centroids)
```

```python
colors=["g.","r.","b."]

for i in range(len(X)):
    plt.plot(X[i][0],X[i][1],colors[labels[i]],markersize=10)
    
plt.scatter(centroids[:,0],centroids[:,1],marker="X",s=150,zorder=10)
plt.show()
```
